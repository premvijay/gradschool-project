```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from astropy import units, constants
```

```python
def gm(om0, ob0, h):
    return om0*h*np.exp(-ob0*(1. + np.sqrt(2.*h)/om0))


def q1(k, om0, ob0, h):
    return k/(gm(om0, ob0, h) * h *units.Mpc**-1) 


def Tbbks(k, om0, ob0, h):
    return (np.log(1. + 2.34*q1(k, om0, ob0, h))/(2.34*q1(k, om0, ob0, h)))*(1.+3.89*q1(k, om0, ob0, h) + (16.1*q1(k, om0, ob0, h))**2. + (5.46*q1(k, om0, ob0, h))**3. + (6.71*q1(k, om0, ob0, h))**4)**(-0.25)
```

```python
h = 0.5
om0 = 1.0
ob0 = 0
k = np.logspace(-3,1,300) * h * units.Mpc**-1
T = Tbbks(k,om0,ob0,h)
```

```python
plt.plot(k.to(h * units.Mpc**-1),T)
plt.xscale('log')
plt.yscale('log')
plt.title("BBKS Transfer Function")
```

```python
H0 = h * 100 * units.km/units.s / units.Mpc
delta_H = 2e-5
factor = 2*np.pi**2 * (constants.c/ H0)**4 * delta_H**2
factor.to(units.Mpc**4)
```

```python
P = (T**2 * k* factor).decompose() # * 6e5 * units.Mpc**4
```

```python
# P.to(h**-3 * units.Mpc**3)
```

```python
plt.figure(dpi=120)
plt.plot(k.to(h * units.Mpc**-1),P.to(h**-3 * units.Mpc**3))
plt.xscale('log')
plt.yscale('log')
plt.xlabel('$k ~($h Mpc$^{-1})$')
plt.ylabel('$P(k)$ (h$^{-3}$ Mpc$^3$)')
plt.title('Matter power spectrum in a standard CDM model ($\Omega_m=${} & $h = ${})'.format(om0,h))
```

```python
plt.figure(dpi=120)
Del = (k**3 * P / (2*np.pi**2) ).decompose()
plt.plot(k.to(h * units.Mpc**-1),Del)
plt.xscale('log')
plt.yscale('log')
plt.xlabel('$k ~($h Mpc$^{-1})$')
plt.ylabel('$\Delta(k)$')
plt.title('Matter power spectrum in a standard CDM model ($\Omega_m=${} & $h = ${})'.format(om0,h))
```

```python
from scipy.special import spherical_jn
```

```python
R_8 = 8 * h**-1 * units.Mpc
def Ws(k):
    k_8 = np.array((k*R_8).decompose())
    return 3* spherical_jn(1,k_8)/k_8
```

```python
Integ = P*k**2*Ws(k)
plt.figure(dpi=120)
plt.plot(k,Integ)
plt.xscale('log')
plt.subplots_adjust(-0.4,-0.3)
plt.title("Just to show that the integrand is negligible outside the range we have choosen.")
```

```python
sigma_8 = np.sqrt(np.trapz(Integ[:-20],k[:-20]))
sigma_8.decompose()
```

```python
# A small demo to find fourier transform of spherical top hat
l = np.linspace(1e-10,30,128)
K = np.array(np.meshgrid(l,l,l))
k = np.sqrt((K**2).sum(axis=0))
gk = spherical_jn(1,5*k)/ (5*k)
plt.imshow(gk[50])
```

```python
gr = np.fft.ifftn(gk)
plt.imshow(np.real(gr[127]))
```
