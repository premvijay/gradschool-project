```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from astropy import units, constants
```

```python
def gm(om0, ob0, h):
    return om0*h*np.exp(-ob0*(1. + np.sqrt(2.*h)/om0))


def q1(k, om0, ob0, h):
    return k/(gm(om0, ob0, h) * h *units.Mpc**-1) 


def Tbbks(k, om0, ob0, h):
    return (np.log(1. + 2.34*q1(k, om0, ob0, h))/(2.34*q1(k, om0, ob0, h)))*(1.+3.89*q1(k, om0, ob0, h) + (16.1*q1(k, om0, ob0, h))**2. + (5.46*q1(k, om0, ob0, h))**3. + (6.71*q1(k, om0, ob0, h))**4)**(-0.25)
```

```python
h = 0.5
om0 = 1.0
ob0 = 0
k = np.logspace(-3,1) * h * units.Mpc**-1
T = Tbbks(k,om0,ob0,h)
```

```python
plt.plot(k.to(h * units.Mpc**-1),T)
plt.xscale('log')
plt.yscale('log')
plt.title("BBKS Transfer Function")
```

```python
H0 = h * 100 * units.km/units.s / units.Mpc
delta_H = 2e-5
factor = 2*np.pi**2 * (constants.c/ H0)**4 * delta_H**2
factor.to(units.Mpc**4)
```

```python
P = T**2 * k* factor # * 6e5 * units.Mpc**4
```

```python
# P.to(h**-3 * units.Mpc**3)
```

```python
plt.figure(dpi=120)
plt.plot(k.to(h * units.Mpc**-1),P.to(h**-3 * units.Mpc**3))
plt.xscale('log')
plt.yscale('log')
plt.xlabel('$k ~($h Mpc$^{-1})$')
plt.ylabel('$P(k)$ (h$^{-3}$ Mpc$^3$)')
plt.title('Matter power spectrum in a standard CDM model ($\Omega_m=${} & $h = ${})'.format(om0,h))
```
